```python
import kagglehub

# Get the data locally:
path = kagglehub.dataset_download("akhiljethwa/blood-cancer-image-dataset", output_dir='./.cache')
print("Path to dataset files:", path)
```

```python
import numpy as np
from pathlib import Path
from skimage.io import imread
from multiprocessing import Pool, cpu_count

# Find all the files:
path = Path(path)
try:
    files_list = [file for file in path.rglob('*.tiff')]
    print(len(files_list), 'files found.')
except Exception as e:
    print('Error finding files:', e)

# Extract image data into a NumPy array:
n_processes = min(4, cpu_count() // 2)
with Pool(processes=n_processes) as pool:
    img_data_arr = pool.map(imread, files_list)

# Create an master NumPy array (n_images x h x w x n_channels)
img_data_arr = np.array(img_data_arr)
print('Image data array shape:', img_data_arr.shape)
np.save('./.cache/Blood_Cancer.npy', img_data_arr)
```

The NumPy array of all the images shows that there are four (4) "layers" of data for each image. I'm guessing this is a CMYK structure, but don't really know how to confirm.

```python
# Example of the layers of the image:
for layer in range(4):
    print(f'\nLayer {layer}:\n', img_data_arr[0, :, :, layer])
```
